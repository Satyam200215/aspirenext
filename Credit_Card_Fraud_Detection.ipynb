```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_data= pd.read_csv('/content/credit_card.csv')
```

```python
#first 5 rows and coloumns of dataset.
#0->legit
#1->fradulent transaction.
credit_data.head()
```

```python
#last 5 rows
credit_data.tail()
```

```python
#info about dataset
credit_data.info()
```

```python
#checking number of missing values in each col
credit_data.isnull().sum()
```

```python
credit_data.isnull().sum()
```

```python
#distribution of transactions
credit_data['Class'].value_counts()
```

```python
#seperating the data for analysis
legit=credit_data[credit_data.Class==0]
fraud=credit_data[credit_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for the both legit and fraud
credit_data.groupby('Class').mean()
```

```python
legit_sample=legit.sample(n=52)
```

```python
#concat two dataframes
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
#splitting into features & target
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split the data into training and test data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Model Training(Logistic Regression Model)
model=LogisticRegression()
```

```python
#training the logisticRegression model
model.fit(X_train,Y_train)
```

```python
#Evaluation(Based on accuracy score)
X_train_prediction=model.predict(X_train)
training_data_acuuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print(training_data_acuuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print(test_data_accuracy)
```
